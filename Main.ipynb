```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./Data/DatasetFinal.csv')
data
```

```python
label_encoder = LabelEncoder()
data['roleId'] = label_encoder.fit_transform(data['role'])
data['rating'] = data['total_mark']
data['skill_set'] = data['skill_set'] / 5
data['development'] = data['development'] / 5
data['culture'] = data['culture'] / 5
data['outstanding'] = data['outstanding'] / 5
data['attitude'] = data['attitude'] / 5
data['employee_skill'] = data['skill_set'] + data['development'] + data['culture'] + data['outstanding'] + data['attitude']
```

```python
data
```

```python
class Display:
    def draw(df,x,y):
        fig = px.scatter_3d(df,x=x,y=y,z='rating',color=x,
                            size='rating',title=f"Employee Rating Matrix" )
        fig.show()
```

```python
Display.draw(data,x='role', y='total_join_year')
```

```python
features = ['roleId', 'skill_set', 'development' , 'culture' , 'outstanding' , 'attitude']
target_variable = 'rating'
```

```python
X_train = data[features]
y_train = data[target_variable]
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42, shuffle=True)
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42, shuffle=True)
```

```python
X_temp
```

```python
y_temp
```

```python
model = Sequential([
    Dense(128, input_shape=(6,), activation='relu'),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(1, activation='linear') 
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=64, validation_data=(X_valid, y_valid))
```

```python
predicted_rating = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, predicted_rating)
print("R-squared:", r_squared)
print("Accuracy:", str(r_squared*100) + "%")
```

```python
skill_set = 5/5
development = 1/5
culture = 1/5
outstanding = 1/5
attitude = 1/5
for i in range(0, 39):
    role_name = label_encoder.inverse_transform([i])
    input_data = np.array([[i,skill_set,development,culture,outstanding,attitude]])
    predicted_rating_new = model.predict(input_data)
    print("Predicted rating" , predicted_rating_new, role_name)
```
